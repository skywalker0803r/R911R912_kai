```python
import pandas as pd
path = "R911R912 _ARO1_R2.xlsx"
xls = pd.ExcelFile(path)
sheets = xls.sheet_names
display(sheets)
```

```python
df = xls.parse(sheets[1])
df.columns.values[2] = "time"
df = df.iloc[:,2:]
for i in df.columns[1:]:
  df[i] = pd.to_numeric(df[i],errors='coerce')

df
```

```python
x_col = df.columns[df.columns.str.contains('_L')].tolist()+df.columns[df.columns.str.contains('FA')].tolist()
x_col = x_col + ["ARO1-DCS-AC_HF"]
```

```python
df_1 = xls.parse(sheets[2])
# df_1 = df_1.iloc[:,1:]
for i in df_1.columns[1:]:
  df_1[i] = pd.to_numeric(df_1[i],errors='coerce')

df_1
```

```python
y_col= ["ARO1-LIMS-s704-744-p-Xylene", "ARO1-LIMS-s912-744-p-Xylene", "ARO1-LIMS-s912-744-m-Xylene",
    "ARO1-LIMS-s917-744-p-Xylene", "ARO1-LIMS-s916-744-p-Xylene", "ARO1-DCS-PX 回收率"]

col = x_col + y_col

df_2 = df[x_col].join(df_1[y_col])
```

```python
# Function to calculate missing values by column# Funct 
def missing_values_table(df):
  # Total missing values
  mis_val = df.isnull().sum()
  
  # Percentage of missing values
  mis_val_percent = 100 * df.isnull().sum() / len(df)
  
  # Make a table with the results
  mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
  
  # Rename the columns
  mis_val_table_ren_columns = mis_val_table.rename(
  columns = {0 : 'Missing Values', 1 : '% of Total Values'})
  
  # Sort the table by percentage of missing descending
  mis_val_table_ren_columns = mis_val_table_ren_columns[
      mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
  '% of Total Values', ascending=False).round(1)
  
  # Print some summary information
  print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
      "There are " + str(mis_val_table_ren_columns.shape[0]) +
        " columns that have missing values.")
  
  # Return the dataframe with missing information
  return mis_val_table_ren_columns

# Missing values statistics
missing_values = missing_values_table(df_2)
missing_values.head(20)
```

```python
# remove outliner
import numpy as np
for j in col[:-1]:
  mean = df_2[j].mean()
  std = df_2[j].std()
  outliner = []
  for i in range(len(df_2[j])):
    if (df_2.loc[i,j] > (mean - std*3) and df_2.loc[i,j] < (mean + std*3)):
      pass
    else:
      outliner = outliner + [i]

  df_2.loc[outliner, j] = np.nan
```

```python
df_3 = df_2[x_col].dropna(axis="index").join(df_2.loc[df_2[x_col].dropna(axis="index").index.tolist(), y_col])
```

```python
import statistics as stat
print("df_3:")
for c in df_3.columns.tolist()[:11]:
    varianceNb = stat.variance(df_3[c].dropna(axis="index"))
    print(c,'變異數 : ', varianceNb)
```

```python
from sklearn.feature_selection import VarianceThreshold
#閥值(threshold)為0.16,表示其變異程度低於0.16就會被removed
print("df_3")
variance_x = VarianceThreshold(threshold = 0.16)
X_ss_vt=variance_x.fit_transform(df_3.iloc[:,:11].dropna(axis="index"))
X_ss_vt=pd.DataFrame(X_ss_vt)
X_ss_vt.head()
```

<font size = 5>Univariate feature selection</font>

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, f_regression

#Suppose, we select 5 features with top 5 Fisher scores
selector = SelectKBest(f_regression, k = 5)
#New dataframe with the selected features for later use in the classifier. fit() method works too, if you want only the feature names and their corresponding scores
for i in y_col[:6]:
  X_new = selector.fit_transform(df_3.loc[df_3[x_col and i].notna(), x_col].dropna(axis="index"), df_3.loc[df_3[x_col and i].notna(), x_col + [i]].dropna(axis="index")[i])
  names = df_3.loc[df_3[i].notna(),x_col].columns.values[selector.get_support()]
  scores = selector.scores_[selector.get_support()]
  names_scores = list(zip(names, [i for j in range(5)], scores))
  ns_df = pd.DataFrame(data = names_scores, columns=['Feat_names', "target_name",'F_Scores'])
  #Sort the dataframe for better visualization
  if i == y_col[0]:
    ns_df_sorted = ns_df.sort_values(['F_Scores', 'Feat_names'], ascending = [False, True])
  else:
    ns_df_sorted = ns_df_sorted.append(ns_df.sort_values(['F_Scores', 'Feat_names'], ascending = [False, True]))

x_new = ns_df_sorted.reset_index(inplace = False)
x_new.iloc[:,1:]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression #F-value between label/feature for regression tasks.

print(df_3.loc[df_3[y_col[0]].notna(),x_col].shape)
model = SelectKBest(f_regression,3)
model.fit_transform(df_3.loc[df_3[x_col and y_col[0]].notna(), x_col].dropna(axis="index"), df_3.loc[df_3[x_col and y_col[0]].notna(), x_col + [y_col[0]]].dropna(axis="index")[y_col[0]])
res = pd.DataFrame(index = x_col)
res["score"] = 0
for i in range(len(res.index)):
  res.iloc[i]["score"] = model.scores_[i]
res.sort_values(by = "score", ascending = False)

# print(df_2.loc[df_2[y_col[0]].notna(),x_col].shape)
# x_new = SelectKBest(f_regression,3).fit_transform(df_2.loc[df_2[x_col and y_col[0]].notna(), x_col].dropna(axis="index"), df_2.loc[df_2[x_col and y_col[0]].notna(), x_col + [y_col[0]]].dropna(axis="index")[y_col[0]])
# print(x_new.shape)
# x_new = pd.DataFrame(x_new)
# x_new.head()
```

<font size = 5>pearsonr</font>

```python
from scipy.stats import pearsonr
X_COL = []
Y_COL = []
pearsonr_corr_list = []
p_value_list = []
for c in df_3.columns.tolist()[:5]:
  for i in y_col[:6]:
    #pearsonr的輸出分別為pearson相關係數和p-value
    pearsonr_corr ,p_value = pearsonr(df_3.loc[df_2[x_col and i].notna(), c].dropna(axis="index"), df_3.loc[df_3[x_col and i].notna(), x_col + [i]].dropna(axis="index")[i])
    X_COL.append(c)
    Y_COL.append(i)
    pearsonr_corr_list.append(pearsonr_corr)
    p_value_list.append(p_value)
    # print(c, " vs ", i, "pearsonr_corr:" ,pearsonr_corr,'p_value',p_value)

pd.DataFrame({"X":X_COL, "Y":Y_COL, "pearsonr_corr":pearsonr_corr_list, "p_value":p_value_list})
```

Ridge

```python
from sklearn.linear_model import Ridge
from sklearn.feature_selection import SelectFromModel
features_names = df_3[x_col].columns.tolist()
#model fit
for j in y_col[:6]:
  print(j, ":")
  ridge = Ridge()
  ridge.fit(df_3.loc[df_3[x_col and j].notna(), x_col].dropna(axis="index"), df_3.loc[df_3[x_col and j].notna(), x_col + [j]].dropna(axis="index")[j])
  #sort coef_
  coef_list=sorted(zip(map(lambda x: round(x,4), ridge.coef_.reshape(-1)),features_names),reverse=True)
  for i in coef_list:
    print(i)
  print()
```
