```python
import pandas as pd
df_x = pd.read_excel('../data/ADS47/R911R912 _明志蔡教授_R4-ARO2.xlsx',sheet_name='R911R912_TAG相關錶點資料(X)')
df_x
```

```python
a_col  = ["ARO2-DCS-R911_2_A_FA",
    "ARO2-DCS-R911_2_HF",
    "ARO2-DCS-R911_2_L2_A",
    "ARO2-DCS-R911_2_L3_A",
    "ARO2-DCS-R911_2_L4_A",
    "ARO2-DCS-R911_2_XF",
    ]
s_col  = [
    "ARO2-DCS-FI914A2",
    "ARO2-DCS-FI914A3",
    "ARO2-DCS-FI91601",
    "ARO2-DCS-FI91701",
    "ARO2-DCS-FI93201",
    "ARO2-DCS-FI94701",
    "ARO2-LIMS-S708@Br.Index",
    "ARO2-LIMS-S708@A9",
    "ARO2-LIMS-S708@A10+",
    "ARO2-LIMS-S708@Water",
    "ARO2-LIMS-S708@Sulfur",
    "ARO2-LIMS-S905@Water",
    "ARO2-LIMS-S907@Water",
]
```

```python
df_y = pd.read_excel('../data/ADS47/R911R912 _明志蔡教授_R4-ARO2.xlsx',sheet_name='目標值(Y)相關錶點資料')
df_y
```

```python
y_col = []
for i in df_y.columns:
    if 'Unnamed' not in i:
        y_col.append(i)
y_col
```

```python
y_col = ['ARO2-LIMS-s922@MX']
```

```python
x_col = s_col+a_col
df_xy = df_x[x_col].join(df_y[y_col])
from tqdm import tqdm
for i in tqdm(df_xy.columns):
    df_xy[i] = pd.to_numeric(df_xy[i],errors='coerce')
df_xy[:] = df_xy[:].fillna(method='ffill').fillna(method='bfill')
df_xy
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
df_xy[:] = ss.fit_transform(df_xy[:])
df_xy
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
model = SVR(C=42)
X,y = df_xy[x_col],df_xy[y_col]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
model.fit(X_train,y_train)
print('train',model.score(X_train,y_train))
print('test',model.score(X_test,y_test))
```

```python
df_xy[y_col].max()
```

```python
import optuna
import numpy as np
request = {} 
request['output'] = df_xy[y_col].max()
def objective(trial: optuna.Trial):
  Input_df = pd.DataFrame(index = [0],columns = x_col)
  for key in x_col:
    Input_df.loc[0,key] = trial.suggest_uniform(name = key,low = df_xy.loc[:,key].min(),high = df_xy.loc[:,key].max())
  output_df = model.predict(Input_df)
  output = output_df[0]
  total_loss = abs(request['output'] - output)
  return total_loss
study = optuna.create_study()
study.optimize(objective, n_trials=42)
```

```python
study.best_params
```

```python
model.predict(pd.DataFrame(study.best_params,index=[0]))[0]
```

```python
from sklearn.metrics import r2_score
pd.set_option('display.max_rows', 500)
df_answer = pd.DataFrame()
for i in study.best_params:
    df_answer[i] = [study.best_params[i]]
df_answer[y_col] = model.predict(pd.DataFrame(study.best_params,index=[0]))[0]
df_answer[:] = ss.inverse_transform(df_answer[:])
def mape(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
def inverse_transform_y(y):
    y = pd.DataFrame(y).fillna(value=168).values
    t = df_xy.copy()
    t = t.sample(len(y_test))
    t[y_col[0]] = y
    t[:] = ss.inverse_transform(t)
    return t[y_col]
y_pred = model.predict(X_test)
print('test r2_score:',r2_score(inverse_transform_y(y_pred),inverse_transform_y(y_test)))
print('test mape:',mape(inverse_transform_y(y_pred),inverse_transform_y(y_test)))
print('answer:')
df_answer.T
```

```python
import matplotlib.pyplot as plt
pred_value = model.predict(X_test)
real_value = y_test.values.reshape(-1)
pred_value = inverse_transform_y(pred_value).values
real_value = inverse_transform_y(real_value).values
plt.figure(figsize=(20,5))
plt.plot(pred_value,label='pred')
plt.plot(real_value,label='real')
plt.legend()
plt.show()
```
