```python
import pandas as pd
df_x = pd.read_excel('data/ADS47/R911R912 _明志蔡教授_R4-ARO2.xlsx',sheet_name='R911R912_TAG相關錶點資料(X)')
df_x
```

```python
a_col  = ["ARO2-DCS-R911_2_A_FA",
    "ARO2-DCS-R911_2_HF",
    "ARO2-DCS-R911_2_L2_A",
    "ARO2-DCS-R911_2_L3_A",
    "ARO2-DCS-R911_2_L4_A",
    "ARO2-DCS-R911_2_XF",
    ]
s_col  = [
    "ARO2-DCS-FI914A2",
    "ARO2-DCS-FI914A3",
    "ARO2-DCS-FI91601",
    "ARO2-DCS-FI91701",
    "ARO2-DCS-FI93201",
    "ARO2-DCS-FI94701",
    "ARO2-LIMS-S708@Br.Index",
    "ARO2-LIMS-S708@A9",
    "ARO2-LIMS-S708@A10+",
    "ARO2-LIMS-S708@Water",
    "ARO2-LIMS-S708@Sulfur",
    "ARO2-LIMS-S905@Water",
    "ARO2-LIMS-S907@Water",
]
```

```python
df_y = pd.read_excel('data/ADS47/R911R912 _明志蔡教授_R4-ARO2.xlsx',sheet_name='目標值(Y)相關錶點資料')
df_y
```

```python
y_col = []
for i in df_y.columns:
    if 'Unnamed' not in i:
        y_col.append(i)
y_col
```

```python
# 選一個示範使用
y_col = ['ARO2-LIMS-s922@MX']
```

```python
df_xy = df_x[s_col+a_col].join(df_y[y_col])
df_xy[s_col+a_col] = df_xy[s_col+a_col].fillna(method='bfill')
df_xy
```

```python
from tqdm import tqdm
for i in tqdm(df_xy.columns):
    df_xy[i] = pd.to_numeric(df_xy[i],errors='coerce')
```

```python
from sklearn.preprocessing import MinMaxScaler

mm_a = MinMaxScaler()
mm_s = MinMaxScaler()
mm_y = MinMaxScaler()

df_xy[a_col] = mm_a.fit_transform(df_xy[a_col])
df_xy[s_col] = mm_s.fit_transform(df_xy[s_col])
df_xy[y_col] = mm_y.fit_transform(df_xy[y_col])
df_xy
```

```python
from tqdm import tqdm_notebook as tqdm
def build_dataset(df,n=8):
    data = {
        'state':[],
        'action':[],
        'value':[],
        'idx':[],
        }
    
    for i in tqdm(range(0,len(df)-n,n)):
        curr = df.iloc[i:i+n] #一小段資料
        cond1 = np.isnan(curr[y_col].values[-1]).sum() == 0
        cond2 = np.isnan(curr[s_col+a_col].values).sum() == 0
        if cond1 & cond2: #確保這段資料無缺失值
            data['state'].append(curr[s_col].values) # 這段時間所有狀態
            data['action'].append(curr[a_col].values) # 這段時間所有操作
            data['value'].append(curr[y_col].values[-1]) # 這段時間最後一個時間點的產量
            data['idx'].append(curr.index[-1]) #這段時間最後 出料對應的時間點
    
    # reshape
    data['state'] = np.array(data['state'])
    data['action'] = np.array(data['action'])
    data['value'] = np.array(data['value'])
    
    # store mm_scaler
    data['mm_state'] = mm_s
    data['mm_action'] = mm_a
    data['mm_value'] = mm_y
    
    # store col_name
    data['state_col'] = s_col
    data['action_col'] = a_col
    data['value_col'] = y_col
    
    return data
```

```python
import numpy as np
data = build_dataset(df_xy.iloc[df_xy[y_col].dropna(axis=0).index[0]+1:],n=8)
```
